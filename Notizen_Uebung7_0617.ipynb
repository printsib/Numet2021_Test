## 2. Exceptions  
Um Fehler abzufangen bzw das Program nicht abstuerzen zu lassen wenn Fehler auftauchen kann man "Exceptions" verwenden. Diese werden als `try-except` statements umgesetzt. Das sieht dann so aus:

```
try: 
    Code das probiert werden soll.
except <Name_von_Exception>:
    Was zu tun ist wenn es scheitert.
```

*Links:*  
https://pythonbasics.org/try-except/

````python
"""
Powermethod implementieren mit einer Exception
"""

def find_eigenvector(A,v):
    """
    Finde Eigenvektor. Abbruchkriterium ist ein Maximaler Durchlaufzaehler
    
    Parameters
    ----------
    A: nxn numpy array
    v: n numpy array
    
    Returns
    -------
    out: list containing eigenvector and eigenvalue
    """
    
    try: 
        A*v
    except ValueError:
        print("Kann A shape ", A.shape, "nicht mit v shape",  v.shape, " multiplizieren. \n Verwende default Werte.")
        # Set default values
        A = np.identity(4)
        v = np.ones(4)
    
    c = 0    # zaehler
    while c < 10e2:
        v_kringel = A.dot(v)
        v = v_kringel / np.linalg.norm(v_kringel)
        eigenval = A.dot(v).dot(v) /v.dot(v)
        c += 1
    
    out = [v, eigenval]
    return out
````

````python
A = np.zeros([6,6])
A[[1,1,1,2,3,4,5,5],[0,3,4,1,0,3,1,4]] = 1/2
A[[0,3,5],2] = 1/3
A[2,5] = 1

# Einen array mit denselben Werten initialisieren.
v = np.full(6, 1/6)
w = np.full(5, 1/6)
````
